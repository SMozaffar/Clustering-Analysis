```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans, SpectralClustering
from sklearn.metrics import pairwise_distances_argmin_min 
import numpy as np
import seaborn as sns

plt.style.use("dark_background")
```

### 1.0 Load the Data and Visualize

```python
# Read in the data as a pandas dataframe
circles = pd.read_csv('ConcentricCircles.csv')

# Make a copy of dataframe to avoid changing original data
df = circles.copy()

# Drop the first column
df = df.drop(df.columns[0], axis=1)

# Extract the x and y coordinates
x = df['x']
y = df['y']

# Create a scatter plot
plt.figure(figsize=(12, 10))
plt.scatter(x, y, alpha=1, s=300, edgecolors='k') 
plt.title('Concentric Circles')
plt.axis('off')

# Display the plot
plt.show()
```

### 2.0 K-means Clustering

```python
# Make a copy of dataframe to avoid changing original data
df = circles.copy()

# Specify the number of clusters
num_clusters = 2

# Perform K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, init='random', random_state=0)
df['cluster'] = kmeans.fit_predict(df[['x', 'y']])

# Set up the color palette
palette = sns.color_palette('bright', num_clusters)

 # Create a scatterplot of the data with clusters indicated by color using seaborn
plt.figure(figsize=(12,10))
sns.scatterplot(data=df, x='x', y='y', hue='cluster', palette=palette, legend=None, s = 300)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:,1], s=600, color='white', marker='X', label='Centroids') 
plt.title('K-Means Clustering with K = 2')
plt.axis('off')
plt.legend()
plt.show()
```

### 2.1 K-means Clustering with Initialalized Means

```python
# Make a copy of dataframe to avoid changing original data
df = circles.copy()

# Specify the number of clusters
num_clusters = 2

# Specify the initial cluster means
initial_centroids = np.array([[-8, 0], [1, -1]])

# Perform K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, init=initial_centroids, n_init=1)
df['cluster'] = kmeans.fit_predict(df[['x', 'y']])

# Set up the color palette
palette = sns.color_palette('bright', num_clusters)

 # Create a scatterplot of the data with clusters indicated by color using seaborn
plt.figure(figsize=(12,10))
sns.scatterplot(data=df, x='x', y='y', hue='cluster', palette=palette, s = 300, legend=None)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:,1], s=600, color='white', marker='X', label='Centroids') 
plt.title('K-Means with Initialized Centroids and K = 2')
plt.axis('off')
plt.legend()
plt.show()
```

### 2.2 K-means Clustering with K = 5

```python
# Make a copy of dataframe to avoid changing original data
df = circles.copy()

# Specify the number of clusters
num_clusters = 5

# Specify the initial cluster means
initial_centroids = np.array([[-7.5, 0], [0, 7.5], [7.5, 0], [0, -7.5], [0, 0]])

# Perform K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, init=initial_centroids, n_init=1)
df['cluster'] = kmeans.fit_predict(df[['x', 'y']])

# Set up the color palette
palette = sns.color_palette('bright', num_clusters)

 # Create a scatterplot of the data with clusters indicated by color using seaborn
plt.figure(figsize=(12,10))
sns.scatterplot(data=df, x='x', y='y', hue='cluster', palette=palette, legend=None, s = 300)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:,1], s=600, color='white', marker='X', label='Centroids') 
plt.title('K-Means Clustering with K = 5')
plt.axis('off')
plt.legend()
plt.show()
```

### 3.0 Spectral Clustering

```python
# Make a copy of dataframe to avoid changing original data
df = circles.copy()

# Specify the number of clusters
num_clusters = 4

# Perform Spectral Clustering
sc = SpectralClustering(n_clusters=num_clusters, affinity='rbf', gamma=1.0/1.5**2)
df['cluster'] = sc.fit_predict(df[['x', 'y']]) 

# Set up the color palette
palette = sns.color_palette('bright', num_clusters)

# Create a scatterplot of the data with clusters indicated by color using seaborn
plt.figure(figsize=(12,10))
sns.scatterplot(data=df, x='x', y='y', hue='cluster', palette=palette, legend=None, s = 300) 
plt.title('Spectral Clustering') 
plt.axis('off')
plt.legend()
plt.show()
```
